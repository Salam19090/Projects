```python
def fnrn(a):
    ans = -1
    for i in a:
        if a.count(i) == 1:
            ans = i
            break
    return ans
```

```python
def firstnonrepeating(a):
    freq = {}
    ans = -1
    for elem in a:
        freq[elem] = a.count(elem)
        
    for elem in a:
        if freq[elem] == 1:
            ans = elem
            break
    return ans
```

```python
a = [9,7,4]
fnrn(a)
```

```python
a = [9,4,9,6,7,4]
firstnonrepeating(a)
```

```python
def missingnumber(a):
    l = len(a)
    n = l+1
    ans = None
    a.sort()
    for i in range(n):
        if i != a[i]-1:
            ans = i +1
            break
    return ans
```

```python
a = [1,3,2,5,6,7,8]
missingnumber(a)
```

```python
def mnumber(a):
    l = len(a)
    n = l+1
    ans = sum(list(range(1,n+1)))-sum(a)
    
    return ans
```

```python
a = [1,3,2,5,6,7,8]
mnumber(a)
```

```python
import numpy as np
def num(a):
    b = np.array(a)
    l = len(a)
    n = l+1
    ans = np.sum(n*(n+1)/2) - np.sum(b)
    return ans
```

```python
a = [1,3,4,5,6,2,9]
num(a)
```

```python
def print_diamond_pattern(size):
    if size % 2 == 0:
        size += 1  # Make sure size is odd for a symmetrical diamond

    # Upper half of the diamond
    for i in range(1, size + 1,2):
        spaces = (size - i) // 2
        stars = i
        print(" " * spaces + "*" * stars)

    # Lower half of the diamond
    for i in range(size - 2, 0, -2):
        spaces = (size - i) // 2
        stars = i
        print(" " * spaces + "*" * stars)
```

```python

import pandas as pd
import numpy as np
def solve(df):
    '''
    input: pandas dataframe
    output: pandas series
    '''
    df = pd.DataFrame({'Date':["2015-12-06", "2011-12-27", "2015-09-07", "2012-12-21", "2020-02-13", "2015-06-09"], 
        'R_id':[498, 721, 375, 464, 813, 853], 
        'Phy':[22, 45, 1, 65, 22, 17], 
        'Chem':[52, 56, 32, 50, 24, 61], 
        'Math':[63, 37, 68, 62, 43, 42]})

    #converting to date time
    df['Date']= pd.to_datetime(df['Date'])

    #extracting month
    df['Month']=df['Date'].dt.month_name()

    #getting value counts and using the highest registration month
    highest_freq = df['Month'].value_counts().values[0]

    #sorting value counts
    month=df['Month'].value_counts().sort_index()

    #getting the frequency of registrations of highest freq month
    freq_month=month[month.values==highest_freq].index[0]

    #calculating average scores of highest freq month
    c = np.round(df[df['Month'] == freq_month]['Chem'].mean(),2)
    p = np.round(df[df['Month'] == freq_month]['Phy'].mean(),2)
    m = np.round(df[df['Month'] == freq_month]['Math'].mean(),2) 

    #making series and returning
    print(pd.Series([freq_month[:3].upper(), highest_freq, c, p, m,]))
```

```python
solve(df)
```

```python
import pandas as pd

# Sample input data
data = {'dates': ["2015-12-06", "2011-12-27", "2015-09-07", "2012-12-21", "2020-02-13"],
        'RID': [498, 721, 375, 464, 813],
        'Phy': [22, 45, 1, 65, 22],
        'Chem': [52, 56, 32, 50, 24],
        'Math': [63, 37, 68, 62, 43]}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the "dates" column to datetime
df['dates'] = pd.to_datetime(df['dates'])

# Filter the DataFrame to select rows on or after May 1, 2014
filtered_df = df[df['dates'] >= '2014-05-01']

# Calculate the average marks for each student
filtered_df['Average'] = filtered_df[['Phy', 'Chem', 'Math']].mean(axis=1)

# Sort the DataFrame by average marks in increasing order
sorted_df = filtered_df.sort_values(by='Average')

# Get the register IDs in the sorted order
result = sorted_df['RID'].tolist()

# Print the result
print(result)
```
