```python
import numpy as np
import matplotlib.pyplot as plt

# Generate an array of 100 evenly spaced values between 0 and 10
x = np.linspace(0, 10, 100)

# Compute the corresponding y values for a simple quadratic function
y = x**2

# Create a plot of x vs. y
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function: y = x^2')
plt.grid(True)
plt.show()
```

```python
import numpy as np

def angle_between_vectors(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    
    cosine_angle = dot_product / (norm_v1 * norm_v2)
    angle = np.arccos(np.clip(cosine_angle, -1, 1))
    
    return angle

def believe_direction(physicist_guess, resident_vector):
    angle = angle_between_vectors(physicist_guess, resident_vector)
    
    if -np.pi/2 <= angle <= np.pi/2:
        return "Believe"
    else:
        return "Do not believe"

# Example vectors
physicist_guess = np.array([1, 0, 0])
resident_vector = np.array([0, 1, 0])
```

```python
believe_direction(np.array([1, 0, 0]), np.array([0, 1, 0]))
```

```python
import numpy as np

a = np.array([4, 1, 1])
b = np.array([2, -4, -4])
x = np.array([1, 1, 4])

coeff_matrix = np.vstack((a, b)).T
constants = np.linalg.solve(coeff_matrix, x)

alpha, beta = constants
print(f"Alpha: {alpha}, Beta: {beta}")
```
