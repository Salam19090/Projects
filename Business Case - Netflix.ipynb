# Handling missing data - 

Following columns does not require fills :-

 *  show_id
 *  type
 *  title
 *  release_year
 *  listed_in
 *  description

Following columns require fills :-

 * director
 * cast
 * country
 * rating
 * duration

To fill the data in missing place will take mode of available values except duration column as missing values of duration column 
misplaced in rating column. 
 * Duration data input error.

```python
data[data.duration.isnull()]
```

As we can see that in above code only director "Louis C.K" has missing values in rating columns. So first need to verify how many rows under director "Louis C.K" . 

```python
data[data.director == 'Louis C.K.'] # This shows total number of rows containing Louis C.K.
```

```python
data.loc[data['director'] == 'Louis C.K.', 'duration'] = data['rating']
data[data.director == 'Louis C.K.']
```

# Heatmap of the given dataset

```python
counts = data.groupby(['type', 'listed_in']).size().unstack()
counts = counts.div(counts.sum(axis=1), axis=0)
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(counts, cmap='Blues', annot=True, fmt='.9%', ax=ax)
plt.title('Percentage of Movies and TV Shows in Each Genre')
plt.show()
```

```python
plt.figure(figsize = (15,10))
sns.pairplot(data.select_dtypes(include=['float64', 'int64']))
plt.title('Pairplot of Netflix Data')
plt.show()
```
